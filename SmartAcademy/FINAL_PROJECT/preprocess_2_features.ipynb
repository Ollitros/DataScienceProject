# Add to data_combined features from minor tables

# 1) bureau.csv

### 1.1) prev_loan_sum

```python
prev_loan_sum = bureau.groupby('SK_ID_CURR', as_index=False)['SK_ID_BUREAU'].count().rename(columns = {'SK_ID_BUREAU': 'prev_loan_sum'})
prev_loan_sum.head()
```

```python
# Join to the training dataframe
data_combined = data_combined.merge(prev_loan_sum, on = 'SK_ID_CURR', how = 'left')

# Fill the missing values with 0 
data_combined['prev_loan_sum'] = data_combined['prev_loan_sum'].fillna(0)
data_combined.head()
```

### 1.2) credit_prolong

```python
credit_prolong = bureau.groupby('SK_ID_CURR', as_index=False)['CNT_CREDIT_PROLONG'].sum().rename(columns = {'CNT_CREDIT_PROLONG': 'credit_prolong'})
credit_prolong.head()
```

```python
# Join to the training dataframe
data_combined = data_combined.merge(credit_prolong, on = 'SK_ID_CURR', how = 'left')

# Fill the missing values with 0 
data_combined['credit_prolong'] = data_combined['credit_prolong'].fillna(0)
data_combined.head()
```

### 1.3) credit_update_days

```python
credit_update_days = bureau.groupby('SK_ID_CURR', as_index=False)['DAYS_CREDIT_UPDATE'].sum().rename(columns = {'DAYS_CREDIT_UPDATE': 'credit_update_days'})
credit_update_days.head()
```

```python
# Join to the training dataframe
data_combined = data_combined.merge(credit_update_days, on = 'SK_ID_CURR', how = 'left')

# Fill the missing values with 0 
data_combined['credit_update_days'] = data_combined['credit_update_days'].fillna(0)
data_combined.head()
```

### 1.4) AMT_CREDIT_SUM_DEBT

```python
amt_credit_sum_debt = bureau.groupby('SK_ID_CURR', as_index=False)['AMT_CREDIT_SUM_DEBT'].sum().rename(columns = {'AMT_CREDIT_SUM_DEBT': 'amt_credit_sum_debt'})
amt_credit_sum_debt.head()
```

```python
# Join to the training dataframe
data_combined = data_combined.merge(amt_credit_sum_debt, on = 'SK_ID_CURR', how = 'left')

# Fill the missing values with 0 
data_combined['amt_credit_sum_debt'] = data_combined['amt_credit_sum_debt'].fillna(0)
data_combined.head()
```

# 2) installments_payments.csv

```python
installments_payments.head()
```

### 1.1) count_prev_sk

```python
count_prev_sk = installments_payments.groupby('SK_ID_CURR', as_index=False)['SK_ID_PREV'].count().rename(columns = {'SK_ID_PREV': 'count_prev_sk'})
count_prev_sk.head()
```

```python
# Join to the training dataframe
data_combined = data_combined.merge(count_prev_sk, on = 'SK_ID_CURR', how = 'left')

# Fill the missing values with 0 
data_combined['count_prev_sk'] = data_combined['count_prev_sk'].fillna(0)
data_combined.head()
```

### 1.2) payment_diff

```python
installments_payments['PAYMENT_DIFF'] = abs(installments_payments['DAYS_ENTRY_PAYMENT']) - abs(installments_payments['DAYS_INSTALMENT'])
payment_diff = installments_payments.groupby('SK_ID_CURR', as_index=False)['PAYMENT_DIFF'].sum().rename(columns = {'PAYMENT_DIFF': 'payment_diff'})

payment_diff.head()
```

```python
# Join to the training dataframe
data_combined = data_combined.merge(payment_diff, on = 'SK_ID_CURR', how = 'left')

# Fill the missing values with 0 
data_combined['payment_diff'] = data_combined['payment_diff'].fillna(0)
data_combined.head()
```

### 1.3) paid_diff

```python
installments_payments['PAID_DIFF'] = abs(installments_payments['AMT_PAYMENT']) - abs(installments_payments['AMT_INSTALMENT'])
paid_diff = installments_payments.groupby('SK_ID_CURR', as_index=False)['PAID_DIFF'].sum().rename(columns = {'PAID_DIFF': 'paid_diff'})

paid_diff.head()
```

```python
# Join to the training dataframe
data_combined = data_combined.merge(paid_diff, on = 'SK_ID_CURR', how = 'left')

# Fill the missing values with 0 
data_combined['paid_diff'] = data_combined['paid_diff'].fillna(0)
data_combined.head()
```

# 3) POS_CASH_balance

```python
POS_CASH_balance.head()
```

### 3.1) months_mean

```python
POS_CASH_balance['MONTHS_BALANCE'] = abs(POS_CASH_balance['MONTHS_BALANCE'])
months_mean = POS_CASH_balance.groupby('SK_ID_CURR', as_index=False)['MONTHS_BALANCE'].mean().rename(columns = {'MONTHS_BALANCE': 'months_mean'})

months_mean.head()
```

```python
# Join to the training dataframe
data_combined = data_combined.merge(months_mean, on = 'SK_ID_CURR', how = 'left')

# Fill the missing values with 0 
data_combined['months_mean'] = data_combined['months_mean'].fillna(0)
data_combined.head()
```

### 3.2) cnt_ratio

```python
POS_CASH_balance['CNT_INSTALMENT_RATIO'] = POS_CASH_balance['CNT_INSTALMENT_FUTURE'] / POS_CASH_balance['CNT_INSTALMENT']
cnt_ratio = POS_CASH_balance.groupby('SK_ID_CURR', as_index=False)['CNT_INSTALMENT_RATIO'].mean().rename(columns = {'CNT_INSTALMENT_RATIO': 'cnt_ratio'})

cnt_ratio.head()
```

```python
# Join to the training dataframe
data_combined = data_combined.merge(cnt_ratio, on = 'SK_ID_CURR', how = 'left')

# Fill the missing values with 0 
data_combined['cnt_ratio'] = data_combined['cnt_ratio'].fillna(0)
data_combined.head()
```

# 4) previous_application

```python
previous_application.head()
```

### 4.1) credit_diff

```python
previous_application['CREDIT_DIFF'] = abs(previous_application['AMT_APPLICATION'] - previous_application['AMT_CREDIT'])
credit_diff = previous_application.groupby('SK_ID_CURR', as_index=False)['CREDIT_DIFF'].mean().rename(columns = {'CREDIT_DIFF': 'credit_diff'})

credit_diff.head()
```

```python
# Join to the training dataframe
data_combined = data_combined.merge(credit_diff, on = 'SK_ID_CURR', how = 'left')

# Fill the missing values with 0 
data_combined['credit_diff'] = data_combined['credit_diff'].fillna(0)
data_combined.head()
```

### 4.2) credit_ratio

```python
previous_application['CREDIT_RATIO'] = previous_application['AMT_APPLICATION'] / previous_application['AMT_CREDIT']
credit_ratio = previous_application.groupby('SK_ID_CURR', as_index=False)['CREDIT_RATIO'].mean().rename(columns = {'CREDIT_RATIO': 'credit_ratio'})

credit_ratio.head()
```

```python
# Join to the training dataframe
data_combined = data_combined.merge(credit_ratio, on = 'SK_ID_CURR', how = 'left')

# Fill the missing values with 0 
data_combined['credit_ratio'] = data_combined['credit_ratio'].fillna(0)
data_combined.head()
```

### 4.3) credit_annunity

```python
previous_application['CREDIT_ANNUITY'] = previous_application['AMT_CREDIT'] / previous_application['AMT_ANNUITY']
credit_annunity = previous_application.groupby('SK_ID_CURR', as_index=False)['CREDIT_ANNUITY'].mean().rename(columns = {'CREDIT_ANNUITY': 'credit_annunity'})

credit_annunity.head()
```

```python
# Join to the training dataframe
data_combined = data_combined.merge(credit_annunity, on = 'SK_ID_CURR', how = 'left')

# Fill the missing values with 0 
data_combined['credit_annunity'] = data_combined['credit_annunity'].fillna(0)
data_combined.head()
```

### 4.4) down_credit

```python
previous_application['DOWN_CREDIT'] = previous_application['AMT_DOWN_PAYMENT'] / previous_application['AMT_CREDIT']
down_credit = previous_application.groupby('SK_ID_CURR', as_index=False)['DOWN_CREDIT'].mean().rename(columns = {'DOWN_CREDIT': 'down_credit'})

down_credit.head()
```

```python
# Join to the training dataframe
data_combined = data_combined.merge(down_credit, on = 'SK_ID_CURR', how = 'left')

# Fill the missing values with 0 
data_combined['down_credit'] = data_combined['down_credit'].fillna(0)
data_combined.head()
```

### 4.5) prev_app_credit_ratio

```python
previous_application['PREV_APP_CREDIT_ratio'] = (previous_application['AMT_ANNUITY'] * previous_application['CNT_PAYMENT'] / 
                                                 previous_application['AMT_CREDIT'] - 1) / previous_application['CNT_PAYMENT']
prev_app_credit_ratio = previous_application.groupby('SK_ID_CURR', as_index=False)['PREV_APP_CREDIT_ratio'].mean().rename(columns = {'PREV_APP_CREDIT_ratio': 'prev_app_credit_ratio'})

prev_app_credit_ratio.head()
```

```python
# Join to the training dataframe
data_combined = data_combined.merge(prev_app_credit_ratio, on = 'SK_ID_CURR', how = 'left')

# Fill the missing values with 0 
data_combined['prev_app_credit_ratio'] = data_combined['prev_app_credit_ratio'].fillna(0)
data_combined.head()
```

### 4.6) prev_days_diff

```python
previous_application['DAYS_DIFF'] = abs(abs(previous_application['DAYS_LAST_DUE_1ST_VERSION']) - 
                                        abs(previous_application['DAYS_LAST_DUE']))
prev_days_diff = previous_application.groupby('SK_ID_CURR', as_index=False)['DAYS_DIFF'].mean().rename(columns = {'DAYS_DIFF': 'prev_days_diff'})

prev_days_diff.head()
```

```python
# Join to the training dataframe
data_combined = data_combined.merge(prev_days_diff, on = 'SK_ID_CURR', how = 'left')

# Fill the missing values with 0 
data_combined['prev_days_diff'] = data_combined['prev_days_diff'].fillna(0)
data_combined.head()
```
