# Upload data

# Find out connections

```python
# Количество уникальных значений займов
data['SK_ID_CURR'].unique().shape
```

```python
# 'SK_ID_CURR' - почти такое же как и в data['SK_ID_CURR']
# 'SK_ID_BUREAU' - это уникальное значения для каждого 'SK_ID_CURR' который несколько раз повторяеться в таблице - 
# типо 'SK_ID_CURR' повторился три раза , а в таблице ему присвоили 'SK_ID_BUREAU' = (1, 2, 3)

print(bureau['SK_ID_CURR'].unique().shape)
print(bureau['SK_ID_BUREAU'].unique().shape)
```

```python
# В документации сказано "Recoded ID of Credit Bureau credit (unique coding for each application) -
# use this to join to CREDIT_BUREAU table". Эту таблицу нужно соеденить с CREDIT_BUREAU, у нас такой таблици нету, 
# а если имееться в виду таблица "bureau", 
# то там и тут не разница между индексами колосальная (см. пункт "Связи по уникальным значениям").
bureau_balance['SK_ID_BUREAU'].unique().shape
```

```python
# installments_payments, POS_CASH_balance, previous_application имеют почти одинаковое количество уникальных займов 'SK_ID_CURR'
# и разное количество предыдущих займов (возможно сдесь логика как и с таблицой bureau)
print(installments_payments['SK_ID_PREV'].unique().shape)
print(installments_payments['SK_ID_CURR'].unique().shape)
```

```python
print(POS_CASH_balance['SK_ID_PREV'].unique().shape)
print(POS_CASH_balance['SK_ID_CURR'].unique().shape)
```

```python
print(previous_application['SK_ID_PREV'].unique().shape)
print(previous_application['SK_ID_CURR'].unique().shape)
```

# Связи по уникальным значениям

```python
# Общие уникальные значения займов из всех таблиц, кроме bureau_balance
SK_ID_CURR = np.intersect1d(data['SK_ID_CURR'].unique(), bureau['SK_ID_CURR'].unique())
SK_ID_CURR = np.intersect1d(SK_ID_CURR, installments_payments['SK_ID_CURR'].unique())
SK_ID_CURR = np.intersect1d(SK_ID_CURR, POS_CASH_balance['SK_ID_CURR'].unique())
SK_ID_CURR = np.intersect1d(SK_ID_CURR, previous_application['SK_ID_CURR'].unique())
SK_ID_CURR.shape
```

```python
# Нету связей по bureau_balance['SK_ID_BUREAU'] и прочими таблицами с 'SK_ID_CURR'
print(np.intersect1d(bureau_balance['SK_ID_BUREAU'].unique(), bureau['SK_ID_CURR'].unique()).shape)
print(np.intersect1d(bureau_balance['SK_ID_BUREAU'].unique(), installments_payments['SK_ID_CURR'].unique()).shape)
print(np.intersect1d(bureau_balance['SK_ID_BUREAU'].unique(), POS_CASH_balance['SK_ID_CURR'].unique()).shape)
print(np.intersect1d(bureau_balance['SK_ID_BUREAU'].unique(), previous_application['SK_ID_CURR'].unique()).shape)
print(np.intersect1d(bureau_balance['SK_ID_BUREAU'].unique(), data['SK_ID_CURR'].unique()).shape)
```

```python
# bureau_balance['SK_ID_BUREAU'] имеет связь с bureau['SK_ID_BUREAU']
print(np.intersect1d(bureau_balance['SK_ID_BUREAU'].unique(), bureau['SK_ID_BUREAU'].unique()).shape)
print(np.intersect1d(bureau_balance['SK_ID_BUREAU'].unique(), installments_payments['SK_ID_PREV'].unique()).shape)
print(np.intersect1d(bureau_balance['SK_ID_BUREAU'].unique(), POS_CASH_balance['SK_ID_PREV'].unique()).shape)
print(np.intersect1d(bureau_balance['SK_ID_BUREAU'].unique(), previous_application['SK_ID_PREV'].unique()).shape)
```
